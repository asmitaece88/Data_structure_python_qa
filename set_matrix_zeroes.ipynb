```python
m , n = len(input_matrix) , len(input_matrix[0])
```

```python
m , n 
```

```python
row , col =[1]*m , [1]*n
```

```python
row , col
```

```python
## update row/col
for  r in range(m):
    for c in range(n):
        if input_matrix[r][c]==0:
            row[r]=0
            col[c]=0
```

```python
# update rows 
for r in  range(m):
     if row[r]==0:
            input_matrix[r]=[0]*n
```

```python
for c in range(n):
    if col[c]==0:
        for i in range(n):
            input_matrix[i][c]=0
```
