# note , this is  reversing , not sorting 

# hence suppose  we have low index and high index , we have to swap these  items 

# next , we have to increment the low index and decrement the high inex 

# keep on incrementing the low index and decremeting the high index 
# until the low index> = high  index

```python
def reverse_num( nums):
    start_index=0
    last_index =len(nums)-1
    
    
    while last_index > start_index:
        nums[start_index], nums[last_index]=nums[last_index], nums[start_index]
        start_index=start_index+1
        last_index=last_index -1
    
    
```

```python
if __name__ =="__main__":
    n_list =[4,56,-90,89,334,15]
    reverse_num(n_list)
    
    print(n_list)
```
