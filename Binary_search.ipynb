# now  if we use binary search algorithm , it uses O(log n )complexity

# there are 2  ways of solving binary search processes 
## iterative or recursive processes 

# iterative process

# pseudo code 

# actual iterative process

```python
def binary_Search_iterative(A,key):
    l=0
    
    r=len(A)-1
    
    while l<=r:
        mid=(l+r)//2
        if key==A[mid]:
            return mid
        elif key< A[mid]:
            r=mid-1
        elif key >A[mid]    :
            l=mid+1
    return -1   
```

```python
A = [15,21,47,84,96]
found = binary_Search_iterative(A,84)+1
found
```

# the complexity is O(log n )

# next we will try the recursive algorithm for binary search 

# the actual code  for binary search - recursion

```python
def binary_Search_recursion(A , key , l , r):
    
    if l >r:
        return -1
    else:
        mid=(l+r)//2
        
        if key==A[mid]:            
            return mid
        elif key< A[mid]:            
            return binary_Search_recursion(A , key,l,mid-1)
        elif key> A[mid] :            
            return binary_Search_recursion(A , key , mid+1 , r)
```

```python
A = [15, 21, 47, 84, 96]
```

```python
found = binary_Search_recursion(A,17,0,4)
print('Result:', found)
```

```python
found = binary_Search_recursion(A,84,0,4)+1
print('Result:', found)
```
