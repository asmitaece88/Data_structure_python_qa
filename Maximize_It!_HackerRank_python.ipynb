```python
## here  the first inout 3 is the number of input values 
## 1000 is the modulo function
from itertools import product 

n,m = [int(x) for x in input().split()]

l1=list()

for i in range(n):
    l=list(map(int , input().split()))[1:]
    #print(l)
    l1.append(l)
r= map(lambda x : sum(i*i for i in x)%m, product(*l1))  
#print(r)
print(max(r))
```
