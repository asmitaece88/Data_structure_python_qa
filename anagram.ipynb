Construct an algorithm to check whether two words (or phrases) are anagrams or not!

"An anagram is a word or phrase formed by rearranging the letters of a different word or phrase, typically using all the original letters exactly once"

For example: restful and fluster

# for checking whetehr the anagram is actually an anagrm of the subject ,  first we need to sort all the letters in an alphabetical
# order and the we  have to iterate thu the  letters and check whether its same or not 

```python
def is_anagram(str1, str2):
    #if leng of string1 not equal to leng of string2 
    #then its not anagram 
    if len(str1)!=len(str2):
        return False
    else:
        # we have to sort the letters of the strings and then we haev to compare
        # the letters with the same indexes
        # this is the bottlenect because it has O(NlogN)
        str1=sorted(str1)
        str2=sorted(str2)
        
        #after that  we have to check  the letters with same indexes 
        #O(n) running time
        for i in range(len(str1)):
            if str1[i]!=str2[i]:
                return False
            
        # overall running time is O(NlogN)+O(N)=O(NlogN)    
        return True     
                
```
